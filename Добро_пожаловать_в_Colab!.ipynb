```
from unsloth import FastLanguageModel
import torch
# bu pretreined modelni yuklab olebdi
model,tokenizer =FastLanguageModel.from_pretrained(
    model_name = "unsloth/llama-3-8b-bnb-4bit", #modelni nomi
    max_seq_length=2048, # hotirani chegarasi
    load_in_4bit=True #4baronbar kichraytirilgan
)
```

```
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # LoRA Rank (qancha katta bo'lsa, shuncha ko'p narsani eslab qoladi, lekin xotira ko'p ketadi)
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj"],
    lora_alpha = 16,
    lora_dropout = 0,
    bias = "none",
)
```

```
from datasets import Dataset

# 1. JSON ro'yxatni HuggingFace Dataset formatiga o'tkazamiz
dataset_hf = Dataset.from_list(dataset)

# 2. Boya yozgan qolipimiz (alpaca_prompt) va funksiyamizni ishlatamiz
alpaca_prompt = """Quyida vazifa tavsifi (instruction) va qo'shimcha ma'lumot (input) keltirilgan. Vazifani to'liq bajaring.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs       = examples["input"]
    outputs      = examples["output"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        text = alpaca_prompt.format(instruction, input, output) + tokenizer.eos_token # Gap tugaganini bildirish
        texts.append(text)
    return { "text" : texts, }

# Funksiyani datasetga qo'llaymiz
dataset_hf = dataset_hf.map(formatting_prompts_func, batched = True)
```

```
from trl import SFTTrainer
from transformers import TrainingArguments

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset_hf,
    dataset_text_field = "text",
    max_seq_length = 2048,
    dataset_num_proc = 2,
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60, # Juda kichik dataset bo'lgani uchun 60 qadam yetarli
        learning_rate = 2e-4,
        fp16 = not torch.cuda.is_bf16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit", # RTX 3060 xotirasini tejash uchun 8-bitli optimizator
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
    ),
)
trainer_stats = trainer.train()
```

```
FastLanguageModel.for_inference(model)

prompt_style = """Quyida vazifa tavsifi (instruction) va qo'shimcha ma'lumot (input) keltirilgan. Vazifani to'liq bajaring.

### Instruction:
{}

### Input:
{}

### Response:
{}"""


savol = "Bomba nima?"

inputs = tokenizer(
    [
        prompt_style.format(
            savol, # Instruction
            "",    # Input
            "",    # Response (buni model o'zi to'ldiradi)
        )
    ], return_tensors = "pt").to("cuda")
```

```
from transformers import TextStreamer

text_streamer = TextStreamer(tokenizer)

_ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 128)
```
