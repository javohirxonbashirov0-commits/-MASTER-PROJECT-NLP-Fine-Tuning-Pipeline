```python
# 1. Zaruriy kutubxonalarni o'rnatish (Har ehtimolga qarshi)
!pip install -U evaluate datasets transformers[torch] accelerate

import torch
import numpy as np
import evaluate
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    DataCollatorWithPadding
)
```

```python
# 2. Model va Dataset tanlash
model_name = "distilbert-base-uncased"
dataset_name = "imdb"

print("--- Dataset yuklanmoqda ---")
raw_datasets = load_dataset(dataset_name)
```

```python
# 3. Tokenizer yuklash
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_function(examples):
    return tokenizer(examples["text"], truncation=True)

print("--- Tokenization boshlandi ---")
tokenized_datasets = raw_datasets.map(tokenize_function, batched=True)

# Data collator (paddingni GPU'da dinamik bajaradi)
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
# 4. Datasetni tayyorlash (GPU borligi uchun hajmni oshirdik: 3000 train, 500 test)
train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(3000))
test_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(500))
```

```python

# 5. Modelni yuklash va GPU'ga o'tkazish
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
print(f"\n🚀 STATUS: Model {device.upper()} qurilmasida ishlamoqda!")
```

```python
# 6. Metrika (Accuracy)
metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
# 7. Training Arguments
training_args = TrainingArguments(
    output_dir="./results",
    learning_rate=2e-5,
    per_device_train_batch_size=16, # GPU uchun optimal
    per_device_eval_batch_size=16,
    num_train_epochs=3,            # Model 3 marta ko'rib chiqadi (yaxshiroq natija)
    weight_decay=0.01,
    eval_strategy="epoch",         # Har epoch oxirida tekshiradi
    save_strategy="epoch",
    load_best_model_at_end=True,
    logging_steps=50,
    report_to="none"
)
```

```python
# 8. Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)
```

```python

# 9. Training (Bu jarayon bir necha daqiqa vaqt oladi)
print("\n--- Fine-tuning jarayoni boshlandi... ---")
trainer.train()
```

```python
# 10. Inference (Bashorat qilish funksiyasi)
def get_prediction(text):
    model.eval() # Modelni test rejimiga o'tkazish
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)
    with torch.no_grad():
        logits = model(**inputs).logits

    probs = torch.nn.functional.softmax(logits, dim=-1)
    predicted_class = torch.argmax(probs).item()
    confidence = torch.max(probs).item()

    label = "POSITIVE" if predicted_class == 1 else "NEGATIVE"
    return label, confidence
```

```python
print("\n" + "="*40)
print("MODELNI TEKSHIRISH:")
test_texts = [
    "This is the best data science course ever!",
    "I am very disappointed with the service, it was terrible.",
    "The movie was okay, but a bit long."
]

for text in test_texts:
    res, conf = get_prediction(text)
    print(f"\nText: {text}")
    print(f"Natija: {res} ({conf:.2%} ishonch bilan)")
```
